```python
a=90.000   #important, adjusted value 
bXR=300.000 #No va 
kXR=3000.000 #typical value 
gXR=0.060 #real, checked 
dXRa=0.1998 #real, important not checked 
bX=300.000 #typical value 
nX=1.000 #Assumed ...why?
kx=1409.148 #not important 
kXRa=3000.000 #esto no va 
gXRa=0.0231 #not important not checked  


aX = 0.004691138 #real, basal checked, important
aR = 0.004691138 #basal Repressor production ---------non validated
gCY= 0.00071666667 #important checked, strange value 
#gaCY = 0.01171666667 # Cytochrome protein (Active) degradation coupling ------ non validated
gaCY = 1171666667
kc = 0.000001406334639 #real, important 
kR = 0.000001406334639 #repressor coupling rate ------ non validated
gO = 0.033 #ompT degradarion rate ------- non validated
gR = 0.033 #Repressor degration rate -------- non validated


#WTFWTFWTFWTFWTFWTFWTFWWTFWTF?!?!?!?!? -----> LEL
maxVal = 300 #deacuerdo a la grafica anterior para CY ------non validated---------MUUUUUST CHECK
hill = 150 #RANDOM VALUE. NO IDEA ------- VERY FLIPPING NON VALIDATED------------MUST CHECK
n = 1.0 #probemos con valores entre 0 y 1. ------non validated

```

```python
#We define the signal function
def signal(t):
    return 5000*sin(2*3.141592*t/30)+4000 #Seno de periodo de media hora
    #return 4000*sin(2*3.141592*t/1440)+4000 #Seno de periodo de un dÃ­a
    #return 0
    
    
    
#We define the step function
def st(t_old, XRa_old, XR_old, CY_old, A_old, R_old, O_old):
    
    XRa = int(XRa_old) #Will be the new  LuxR + AHL
    XR = int(XR_old)   #Will be the new  LuxR
    CY = int(CY_old)   #Will be the new  Cytochrome Protein (Conductivity)
    
    A = int(A_old)     #Will be the new  AHL (will not be used)
    R = int(R_old)     #Will be the new  TetR
    O = int(O_old)     #Will be the new  ompT
    
    ks=[]        #Probability constants
    cltive=[]    #Cumulative constants
    
    #XRa events: kc*XR*A-(gXRa*XRa)-(dXRa*XRa)
    
    k0 = kc*XR*A         #XRa +=1, XR -=1
    k1 = gXRa*XRa        #XRa -=1
    k2 = dXRa*XRa        #XRa -=1, XR +=1
    ks.append(k0)
    ks.append(k1)
    ks.append(k2)
    
    #XR events: a-(kc*A*XR)-(gXR*XR)+(dXRa*XRa)
    
    k3 = a               #XR +=prod
     #k4 = kc*XR*A         #XR -=1, XRa +=1 ------------------------------------------------
    k5 = gXR*XR          #XR -=1
     #k6 = dXRa*XRa        #XR +=1, XRa -=1 ------------------------------------------------
    ks.append(k3)
    ks.append(k4)
    ks.append(k5)
    ks.append(k6)
    
    #CY events: aX+ (termino de acoplo de KRa y CY) -(gCY*CY)-(gaCY*CY*O)
    
    k7 = aX              #CY +=prod
    k8 = whuuuuuuuuuuuuuuuuuuuuuuuuuuuuuut         #CY +=prod, R +=prod 
    k9 = gCY*CY          #CY -=1
    k10 = gaCY*CY*O      #CY -=1 ------------> asumo que el ompT no se destruye al actuar con el citocromo
    ks.append(k7)
    ks.append(k8)
    ks.append(k9)
    ks.append(k10)
    
    #R events: aR+(kc*A*XR)-(gR*R)
    
    k11 = aR             #R +=prod---------> asumo que crece de a 60, tal y como el CY lo hace (por el circuito)
    k12 = kc*A*XR        #R +=prod
    k13 = gR*R           #R -=1
    ks.append(k11)
    ks.append(k12)
    ks.append(k13)
    
    #O events: aX-(kR*R)-(gO*O)
    
    k14 = aX             #O +=1
    k15 = kR*R           #O -=1
    k16 = gO*O           #O -=1
    ks.append(k14)
    ks.append(k15)
    ks.append(k16)
    
    
    #Normalization constant
    s = 0.0          
    for j in ks:
        s = s+j 
    
    #We create the cumulative constants array
    for j in range(len(ks)):
        if(j==0):
            cltive.append(ks[0]/s)
        else:
            zzz = float(ks[j]/s)
            cltive.append(cltive[j-1] + zzz)
     
    #Use the random number r1 to generate the next interesting event
    r1 = random.random()
    T = (1/s)*log(1/r1) + t_old 
    
    #Here we refresh the signal concentration
    A = signal(T)
    #Now we refresh the values choosing a reaction
    prod = 10 #---------------------------------------> Este valor es muy importante, cambia toda la escala... IDK!!!
    ran = random.random()
    
    if(cltive[0]<=ran<cltive[1] and XRa!=0):
        XRa -=1
            
    elif(cltive[1]<=ran<cltive[2] and XRa!=0):
        XRa -=1
        XR +=1
            
    elif(cltive[2]<=ran<cltive[3]):
        XR +=prod
            
    elif(cltive[3]<=ran<cltive[4] and XR!=0):
        XR -=1
        XRa +=1 
            
    elif(cltive[4]<=ran<cltive[5] and XR!=0):
        XR -=1
            
    elif(cltive[5]<=ran<cltive[6] and XRa!=0):
        XR +=1
        XRa -=1
            
    elif(cltive[6]<=ran<cltive[7]):
        CY +=prod
            
    elif(cltive[7]<=ran<cltive[8] and XR!=0):
        CY +=prod
        XR -=1
    
    elif(cltive[8]<=ran<cltive[9] and CY != 0):
        CY -=1
            
    elif(cltive[9]<=ran<cltive[10] and CY != 0):
        CY -=1
            
    elif(cltive[10]<=ran<cltive[11]):
        R +=prod
            
    elif(cltive[11]<=ran<cltive[12]):
        R +=1
            
    elif(cltive[12]<=ran<cltive[13] and R!=0):
        R -=1
            
    elif(cltive[13]<=ran<cltive[14]):
        O +=1
            
    elif(cltive[14]<=ran<cltive[15] and O!=0):
        O -=1
            
    elif(cltive[15]<=ran<cltive[16] and O!=0):
        O -=1
    
    #At last, we return the new values
    return T, XRa, XR, CY, A, R, O
    
    
    
    
    
#We define the Cell function, its input is the time you want to run the system IN HOURS, the window of time for
#standardize the arrays (dt: something between 0.3 and 0.5 should do it), and the initial concentration 
#of each of the components of the system.
#Returns all the arrays of the concentrations (standardized and regular ones)
def cell(hours, dt, XRa_0, XR_0, CY_0, A_0, R_0, O_0):
    
    T = []
    XRa = []
    XR = []
    CY = []
    A = []
    R = []
    O = []    
    
    XRa_0 = int(XRa_0)
    XR_0 = int(XR_0)
    CY_0 = int(CY_0)
    A_0 = int(A_0) 
    R_0 = int(R_0)
    O_0 = int(O_0)
    
    
    t = 0.0
    #add the initial values
    T.append(0.0)
    XRa.append(XRa_0)
    XR.append(XR_0)
    CY.append(CY_0)
    A.append(A_0)
    R.append(R_0)
    O.append(O_0)
    
    #Let's run it!!
    while t < hours*60.0:
        
        #We keep using the same names for the variables, except they are not initial values anymore.
        t, XRa_0, XR_0, CY_0, A_0, R_0, O_0 = st(t, XRa_0, XR_0, CY_0, A_0,R_0,O_0)
        #Saving the simulated values
        T.append(t)
        XRa.append(XRa_0)
        XR.append(XR_0)
        CY.append(CY_0)
        A.append(A_0)
        R.append(R_0)
        O.append(O_0)
        
    #In order to average between a lot of cells, we have to standardize the time.
    #Here we use dt to calculate how many intervals do we want.
    T_s = linspace(0, hours*60.0, int(hours*60.0/dt))
    XRa_s = zeros(len(T_s))
    XR_s = zeros(len(T_s))
    CY_s = zeros(len(T_s))
    A_s = zeros(len(T_s))
    R_s = zeros(len(T_s))
    O_s = zeros(len(T_s))
    
    
    for i in range(len(T_s)):
        j = 0
        while(T[j] < T_s[i]):
            j+=1
            
        XRa_s[i] = XRa[j]
        XR_s[i] = XR[j]
        CY_s[i] = CY[j]
        A_s[i] = A[j]
        R_s[i] = R[j]
        O_s[i] = O[j]
    
    #Finally, we return the arrays. Fisrt the standardized ones.
    
    return T_s, XRa_s, XR_s, CY_s, A_s, R_s, O_s, T, XRa, XR, CY, A, R, O






#We define the Several Cells function, its input is the number of systems you wanto simate, the time you want to run 
#the system IN HOURS, the window of time for standardize the arrays (dt: something between 0.3 and 0.5 should do it), 
#and the initial concentration of each of the components of the system.
#Returns all the arrays of the concentrations
def sv_cells(N_cells, hours, dt, XRa_0, XR_0, CY_0, A_0, R_0, O_0):
    
    #We define the amount of points in the temporal axis
    n_points = int(hours*60.0/dt)
    #We define one matrix for each of the components of the system
    XRa = zeros((N_cells, n_points))
    XR = zeros((N_cells, n_points))
    CY = zeros((N_cells, n_points))
    A = zeros((N_cells, n_points))
    R = zeros((N_cells, n_points))
    O = zeros((N_cells, n_points))
    
    #Now we fill the matrices, running N_cells times
    for i in range(N_cells):
        
        stuff = cell(hours, dt, XRa_0, XR_0, CY_0, A_0, R_0, O_0)
        
        T_s = stuff[0]
        XRa[i,:] = stuff[1]
        XR[i,:] = stuff[2]
        CY[i,:] = stuff[3]
        A[i,:] = stuff[4]
        R[i,:] = stuff[5]
        O[i,:] = stuff[6]
    
    #Initialize the average arrays
    ave_XRa = zeros( n_points)
    ave_XR = zeros( n_points)
    ave_CY = zeros( n_points)
    ave_A = zeros( n_points)
    ave_R = zeros( n_points)
    ave_O = zeros( n_points)
    
    #Fill the average arrays
    for k in range(n_points):
        ave_XRa[k] = average(XRa[:,k])
        ave_XR[k] = average(XR[:,k])
        ave_CY[k] = average(CY[:,k])
        ave_A[k] = average(A[:,k])
        ave_R[k] = average(R[:,k])
        ave_O[k] = average(O[:,k])
    
    #Finally we return the arrays made
    return T_s, ave_XRa, ave_XR, ave_CY, ave_A, ave_R, ave_O


print("lol")
```
